### Categorical to Numerical

```python
binary=[0,1]
na_binary=[0,1,2]
yn= ['Yes', 'No']
ynp=['Yes','No','No phone service']
yni=['Yes','No','No internet service']

tcc['Gender'].replace(['Male','Female'],binary, inplace=True)
tcc['Senior Citizen'].replace(yn,binary,inplace=True)
tcc['Partner'].replace(yn,binary,inplace=True)
tcc['Dependents'].replace(yn,binary,inplace=True)
tcc['Phone Service'].replace(yn,binary,inplace=True)
tcc['Multiple Lines'].replace(ynp,na_binary,inplace=True)
tcc['Internet Service'].replace(['No','Fiber optic', 'DSL'],na_binary,inplace=True)
tcc['Online Security'].replace(yni,na_binary,inplace=True)
tcc['Online Backup'].replace(yni,na_binary,inplace=True)
tcc['Device Protection'].replace(yni,na_binary,inplace=True)
tcc['Tech Support'].replace(yni,na_binary,inplace=True)
tcc['Streaming TV'].replace(yni,na_binary,inplace=True)
tcc['Streaming Movies'].replace(yni,na_binary,inplace=True)
tcc['Contract'].replace(['Month-to-month', 'Two year', 'One year'],na_binary,inplace=True)
tcc['Paperless Billing'].replace(yn,binary,inplace=True)
tcc['Payment Method'].replace(['Electronic check','Mailed check','Bank transfer (automatic)','Credit card (automatic)'],[0,1,2,3],inplace=True)
```
