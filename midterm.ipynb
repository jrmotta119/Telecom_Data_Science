```python
import numpy as np

mu, sigma = 0, 1

gauss = np.random.normal(mu,sigma,1000000)
```

Each Cell is a question.

```python
import matplotlib.pyplot as plt

n_bins = 100
plt.hist(gauss,n_bins)
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Histogram of values in Gauss')
```

```python
meanG = np.mean(gauss)
meanG
```

```python
sdG = np.std(gauss)
sdG
```

```python
cauchy = np.random.standard_cauchy(1000000)
plt.hist(cauchy,n_bins)
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Histogram of values in Cauchy')
```

```python
truncated = cauchy[(cauchy>-10) & (cauchy <10)]
plt.hist(truncated,n_bins)
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Histogram of values in Cauchy')


```

```python
meanC = np.mean(cauchy)
meanC
```

```python
sdC = np.std(cauchy)
sdC
```

```python
def countOutofRange(L, aveL, sdL, k):
    num = 0
    for x in L:
        if (x<(aveL-(k*sdL)) or x >(aveL + (k*sdL))):
            num += 1
    return num
```

```python
Results = np.zeros((10,2))
Results
```

```python
k = [1,2,3,4,5,10,20,50,100,250]
i = 0
for x in Results:
    x[0]=countOutofRange(gauss,meanG,sdG,k[i])
    x[1]=countOutofRange(cauchy,meanC,sdC,k[i])
    i +=1
Results
```

13.This shows the measure of data variability between both distributions. Showing that Gauss has more variability arround the mean with the same k values than Cauchy distribution.
